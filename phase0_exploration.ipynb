```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download NIFTY 50 data (try ^NSEI, if fails use ^NSEBANK or RELIANCE.NS)
nifty = yf.download("^NSEI", start="2015-01-01", end="2026-03-16")

# Plot closing price
plt.figure(figsize=(12,6))

plt.plot(nifty['Close'])
plt.title('NIFTY 50 Closing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
